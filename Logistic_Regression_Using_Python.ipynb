```python
data3 = data3.drop(['StageID','GradeID'], axis = 1)
# as we don't required these columns for prediction of result... so we remove that column from our data
```

```python
data2 = data3[['raisedhands','VisITedResources','AnnouncementsView','Discussion',
               'StudentAbsenceDays','Topic','ParentAnsweringSurvey','ParentschoolSatisfaction','Class']]
# take only those columns which are required remove unneccessary columns.....
```

```python
data2.head()
```

```python
#X = data3.iloc[:,:13]
#y = data3.iloc[:,13:]
X = data2.iloc[:,:8]
y = data2.iloc[:,8:]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn import metrics
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
logistic_reg= LogisticRegression()
logistic_reg.fit(X_train,y_train)
```

```python
y_pred=logistic_reg.predict(X_test)
```

```python
y_pred
```

```python
y_pred_train = logistic_reg.predict(X_train)
```

```python
print(y_pred_train)
print(y_train)
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
print('Accuracy: ',metrics.accuracy_score(y_test, y_pred))
```
